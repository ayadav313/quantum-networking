```python
from qiskit import *
```

```python
circuit1= QuantumCircuit(2,2)
initial_state = [0,1] #Vector representation of |1>
circuit1.initialize(initial_state,0)
circuit1.initialize(initial_state,1)
circuit1.x(0)
circuit1.barrier(0,1)
circuit1.h(0)
circuit1.barrier(0,1)
circuit1.cnot(0,1)
circuit1.barrier(0,1)
circuit1.h(1)
circuit1.barrier(0,1)
circuit1.measure(0,0)
circuit1.measure(1,1)
circuit1.draw()
```

```python
from qiskit.tools.visualization import plot_histogram
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit1, backend=simulator, shots=1000).result()
counts = result.get_counts()
plot_histogram(counts)
```

```python
from qiskit.tools.monitor import job_monitor
IBMQ.load_account()
provider= IBMQ.get_provider('ibm-q')
qcomp = provider.get_backend('ibmq_lima')
job = execute(circuit1, backend=qcomp, shots=1000 )
job_monitor(job)
result= job.result()
print(result.get_counts())
plot_histogram(result.get_counts(circuit1))
```
